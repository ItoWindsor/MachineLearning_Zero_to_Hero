```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
f1 = lambda x : x**2 + 15*np.sin(x)
f2 = lambda x : 2*f1(x)
f1_prime = lambda x : 2*x + 15*np.cos(x)
f2_prime = lambda x : 2*f1_prime(x)
```

```python
x = np.linspace(-5,7)
y1 = f1(x)
t2 = f2(x)
```

```python
x01 = 0 
x02 = 2.1 
alpha = 0.02
xk1 = [x01]
xk2 = [x02]
for k in range(20):
    xk1.append(xk1[-1] - alpha*f1_prime(xk1[-1]))
    xk2.append(xk2[-1] - alpha*f1_prime(xk2[-1]))
xk1 = np.array(xk1)
xk2 = np.array(xk2)
```

```python
plt.plot(x,y1)
plt.scatter(x01,f1(x01), marker = 'x', color = 'black')
plt.scatter(x02,f1(x02), marker = 'x', color = 'black')
plt.scatter(xk1,f1(xk1), marker = 'o', color = 'red', label = f"grad descent : x0 = {x01}", alpha = 0.5)
plt.scatter(xk2,f1(xk2), marker = 'o', color = 'orange', label = f"grad descent : x0 = {x02}", alpha = 0.5)
plt.title("f : x --> x^2 + 15*sin(x)" )
plt.grid()
plt.legend()
plt.show()
```
